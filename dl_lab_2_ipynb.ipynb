```python
#23070521122, Rushi Parhad
import numpy as np
import pandas as pd
```

```python
x=np.array([1,2,3,4,5],dtype=float)
y=np.array([20,40,50,65,80],dtype=float)
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
```

```python
print(x_mean)
print(y_mean)
```

```python
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

slope = numerator / denominator
print(slope)
```

```python
intercept = y_mean - slope * x_mean
print(intercept)
```

```python
print(f"y = {slope:.2f}x + {intercept:.2f}")
```

```python
y_pred = slope * x + intercept
```

```python
for i in range(len(x)):
    print(f"x = {x[i]:.0f}   Predicted y = {y_pred[i]:.2f}")
```

```python
residuals = y - y_pred
for i in range(len(x)):
    print(f"Actual = {y[i]:.0f}   Predicted = {y_pred[i]:.2f}   Residual = {residuals[i]:.2f}")
```

```python
absolute_error = np.abs(residuals)

MAE = np.mean(absolute_error)

print("\nAbsolute Errors =", absolute_error)
print("Mean Absolute Error (MAE) =", MAE)
```

```python
squared_error = residuals ** 2

MSE = np.mean(squared_error)

print("\nSquared Errors =", squared_error)
print("Mean Squared Error (MSE) =", MSE)
```

```python
RMSE = np.sqrt(MSE)

print("\nRoot Mean Squared Error (RMSE) =", RMSE)
```
